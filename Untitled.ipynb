```python
url = "https://api.rootnet.in/covid19-in/unofficial/covid19india.org/statewise/history"
import requests
```

```python
data = requests.get(url)
```

```python
d = data.json()
```

```python
days = d["data"]["history"]
```

```python
allStates = list(map(lambda x: x['state'],days[0]["statewise"]))
```

```python
allDays = list(map(lambda x: x["day"],days))
```

```python
days
```

```python
csv = "name,"+','.join(allDays)+"\n"
```

```python
toMeasure = "confirmed"
for state in allStates:
    stCases = [state]
    for day in days:
        stObj = list(filter(lambda x: x["state"]==state, day["statewise"]))
        print(stObj,len(stObj))
        if len(stObj)<1:
            stCases.append("0")
        else:
            stCases.append(str(stObj[0][toMeasure]))
    stString = ",".join(stCases)
    csv+=stString+"\n"
```

```python
csv
```

```python
with open('file.csv','w') as f:
    f.write(csv)
    f.close()
```
